```python
#Sales analysis project 
import numpy as np
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv("SuperMarket Analysis.csv")
```

```python
# Data cleaning and filling missing values
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.dropna(subset=['Date'])
num = df.select_dtypes(include=['int64', 'float64']).columns
df[num] = df[num].fillna(df[num].median())
df = df.drop_duplicates()
df.head()
```

```python
#feature extraction
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['year'] = df['Date'].dt.year
```

```python
#Finding the beat product
best_products = df.groupby('Product line')['Quantity'].sum().sort_values(ascending=False)
print("\nBest Selling Products:\n", best_products)
```

```python
#monthly revenue calculation
monthly_revenue = df.groupby('Month')['gross income'].sum().sort_values(ascending=False)
print("\nMonthly Revenue:\n", monthly_revenue)
```

```python
# City-wise revenue calculation
city_revenue = df.groupby('City')['gross income'].sum().sort_values(ascending=False)
print("\nCity-wise Revenue:\n", city_revenue)

```

```python
# Payment method usage
payment_counts = df['Payment'].value_counts()
print("\nPayment Method Usage:\n", payment_counts)
```

```python
# Summary
print("\n--- SALES INSIGHT. ")
print("Top Product:", best_products.index[0])
print("Highest Revenue Month:", monthly_revenue.index[0])
print("Top City:", city_revenue.index[0])
print("Most Used Payment Method:", payment_counts.index[0])
```
